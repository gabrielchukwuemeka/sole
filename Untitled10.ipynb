 FIVE PYTHON VISUALIZATIONS FOR SCATTER CHATS

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 35]
sizes = [100, 200, 300, 400, 500]  # Size of the points
colors = [10, 20, 25, 30, 35]  # Color of the points

# Creating the scatter plot
plt.figure(figsize=(10, 6))
scatter = plt.scatter(x, y, s=sizes, c=colors, alpha=0.5, cmap='viridis')

# Adding color bar
plt.colorbar(scatter)

# Adding titles and labels
plt.title('Sample Scatter Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# Displaying the scatter plot
plt.show()
```

 FIVE PYTHON SCRIPTS FOR BOX PLOTS

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
np.random.seed(10)
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Creating box plot
plt.boxplot(data, vert=True, patch_artist=True)

# Adding titles and labels
plt.title('Box Plot Example')
plt.xlabel('Data Groups')
plt.ylabel('Values')

# Displaying the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Creating line chart
plt.plot(x, y, label='Sine Wave')

# Adding titles and labels
plt.title('LINE CHART DISTRIBUTION')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()

# Displaying the plot
plt.show()
```

```python

import matplotlib.pyplot as plt

# Sample data
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice

# Creating pie chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)

# Adding title
plt.title('Pie Chart ')

# Displaying the plot
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
np.random.seed(0)
data = np.random.rand(10, 12)

# Creating heatmap
sns.heatmap(data, annot=True, fmt=".2f", cmap='coolwarm')

# Adding titles and labels
plt.title('Heatmap ')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# Displaying the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Sample data
np.random.seed(0)
data = pd.DataFrame({
    'drinks': np.repeat(['trophy', 'goldberg', 'stout', 'guiness'], 100),
    'Value': np.concatenate([np.random.normal(loc, 0.5, 100) for loc in range(4)])
})

# Creating violin plot
sns.violinplot(x='drinks', y='Value', data=data, palette='muted')

# Adding titles and labels
plt.title('Violin Plot Example')
plt.xlabel('drinks')
plt.ylabel('Value')

# Displaying the plot
plt.show()
```
